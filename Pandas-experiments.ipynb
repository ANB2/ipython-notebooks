```python
import pandas as pd
import numpy

N = 100
data = pd.DataFrame({
    'A': numpy.random.randint(0, 5, N),
    'B': numpy.random.randint(0, 3, N),
    'X': numpy.random.rand(N),
    'Y': numpy.random.rand(N),
})
```

```python
data.groupby(['A', 'B']).agg({'X': ['sum', 'count'], 'Y': ['sum', 'count']})
```
